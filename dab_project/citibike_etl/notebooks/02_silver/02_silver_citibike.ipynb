```python
from pyspark.sql.functions import create_map, lit, current_timestamp, to_date, col
from src.citibike.citibike_utils import get_trip_duration
```

```python
df = spark.read.table("citibike_dev.01_bronze.jc_citibike")
```

```python
df = get_trip_duration(spark, df, "started_at", "ended_at", "trip_duration_mins")
```

```python
df = df.withColumn("trip_start_date", to_date(col("started_at")))
```

```python
df.withColumn("metadata", create_map(lit("pipeline_id"), lit("placeholder"), lit("run_id"), lit("placeholder"), lit("process_timestamp"), current_timestamp()))
```

```python
df = df.select(
    "ride_id",
    "trip_start_date",
    "started_at",
    "ended_at",
    "start_station_name",
    "end_station_name",
    "trip_duration_mins",
    "metadata"
)
```
